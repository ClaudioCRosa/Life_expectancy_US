# U.S. Medical Insurance Costs

```python
import csv

with open("insurance.csv") as insurance:
    insurance_reader = csv.DictReader(insurance, delimiter=",")
    agestr = []
    sex = []
    bmistr = []
    childrenstr = []
    smoker = []
    region = []
    chargesstr = []
    for col in insurance_reader:
        agestr.append(col["age"])
        sex.append(col["sex"])
        bmistr.append(col["bmi"])
        childrenstr.append(col["children"])
        smoker.append(col["smoker"])
        region.append(col["region"])
        chargesstr.append(col["charges"])
    age = [int(age) for age in agestr]
    bmi = [float(bmi) for bmi in bmistr]
    charges = [float(charges) for charges in chargesstr]
    children = [int(children) for children in childrenstr]
    
```

```python
n = len(sex)
average_age = round(sum (age) / len(age),2)
smoking_counter = 0
for answer in smoker:
    if answer == "yes":
        smoking_counter += 1
smoking = round((smoking_counter / len(smoker))*100, 2)
smoker_charges = list(zip(smoker, charges))
paid_smokers = 0
paid_n_smokers = 0
smoker_count = 0
n_smoker_count = 0
for response in smoker_charges:
    if response[0] == "yes":
        paid_smokers += response[1]
        smoker_count += 1
    elif response[0] == "no":
        paid_n_smokers +=response[1]
        n_smoker_count +=1
        
avg_paid_smokers = round(paid_smokers / smoker_count, 2)
avg_paid_n_smokers = round(paid_n_smokers / n_smoker_count, 2)
smoker_ratio = round(avg_paid_smokers/avg_paid_n_smokers * 100, 2)

children_charges = list(zip(children, charges))
max_children = max(children)
paid_childless = 0
childless_count = 0
paid_child1 = 0
childcount1 = 0
paid_child2 = 0
childcount2 = 0
paid_child3 = 0
childcount3 = 0
paid_child4 = 0
childcount4 = 0
paid_child5 = 0
childcount5 = 0
for response in children_charges:
    if response[0] == 0:
        paid_childless += response[1]
        childless_count += 1
    elif response[0] == 1:
        paid_child1 += response[1]
        childcount1 += 1
    elif response[0] == 2:
        paid_child2 += response[1]
        childcount2 += 1
    elif response[0] == 3:
        paid_child3 += response[1]
        childcount3 += 1
    elif response[0] == 4:
        paid_child4 += response[1]
        childcount4 += 1
    elif response[0] == 5:
        paid_child5 += response[1]
        childcount5 += 1

avg_paid_childless = round(paid_childless / childless_count, 2)
avg_paid_child1 = round(paid_child1 / childcount1, 2)
avg_paid_child2 = round(paid_child2 / childcount2, 2)
avg_paid_child3 = round(paid_child3 / childcount3, 2)
avg_paid_child4 = round(paid_child4 / childcount4, 2)
avg_paid_child5 = round(paid_child5 / childcount5, 2)
percent_many_children = round((childcount4 + childcount5+ childcount4)/len(sex) * 100, 2)

sex_charges = list(zip(sex, charges))
paid_men = 0
paid_women = 0
men_counter = 0
women_counter = 0
for response in sex_charges:
    if response[0] == "male":
        paid_men += response[1]
        men_counter += 1
    elif response[0] == "female":
        paid_women += response[1]
        women_counter += 1

avg_paid_men = round(paid_men / men_counter, 2)
avg_paid_women = round(paid_women / women_counter, 2)

sex_smokers = list(zip(sex, smoker))
men_smoker_counter = 0
women_smoker_counter = 0
for response in sex_smokers:
    if response[0] == "male" and response[1] == "yes":
        men_smoker_counter += 1
    if response [0] == "female" and response [1] == "yes":
        women_smoker_counter += 1
        

        
percent_men_smoker = round((men_smoker_counter / men_counter)*100)
percent_women_smoker = round((women_smoker_counter / women_counter)*100)


print("The study included " + str(n)+ " participants whose average age is " + str(average_age) + 
      " years. \nOf them, " + str(smoking) + " percent smoke. \nSmokers pay on average " + str(avg_paid_smokers) + 
      " dollars, whilst non-smokers pay only " + str(avg_paid_n_smokers) + 
      " dollars. \nThis means that by smoking, one increases their insurance costs by " + str(smoker_ratio) + 
      " percent. \nIn other words, tobacco consumption almost quintuples one's insurance costs")
print("The individual with the most children in the sample has " + str(max_children) + 
      " children. \nConcerning insurance costs, the childless pay, on average, " + str(avg_paid_childless) + 
      " dollars. \nThose that do have one, two, three, four or  five children pay, on average, " + 
       str(avg_paid_child1) + " ," + str(avg_paid_child2) + " ," + str(avg_paid_child3) + " ," + str(avg_paid_child4) +
      " ," + str(avg_paid_child5) + " , respectively. \nOne should, however, be cautious when drawing" + 
      " conclusions regarding the effect of having many children on one's insurance costs." + 
      " Those with more than two children constitute only " + str(childcount4 + childcount5 + childcount3) + " individuals, or "
      + str(percent_many_children) + " percent of the sample. \nIf those 200 individuals happen to have a lower incidence of being smokers, " 
      + " \nthe insurance costs to having more than two hildren will be underestimated." + 
      " Therefore, we should disragard them, and  conclude that having children increases an individual's insurance cost," + 
      " though not linearly: the first child increases the costs by around 500 dollars," + " the second child by around 5 times that,"
      " a bigger sample of more individuals with more than two children could either: reaffirm the finding that additional children have a diminishing impact on insurance costs or a positive linear effect, though doubtively ever negatively linear")
print("Genderwise, insurance costs average at " + str(avg_paid_men) + " for men and " + str(avg_paid_women) + " for women." + 
     " This can be due to discrimination, " +
     " \nsuch as a perceived higher risk behaviour in men which leads them to being charged higher costs, " + 
     "or, \ndue to a third variable that was not accounted for. For instance, and as shown above," + 
     " \nthose who smoke have higher insurance costs possibly due to being exposed to higher health risks." + 
     " \nMen also smoke at a higher rate than women: " + str(percent_men_smoker) + " percent, for men, vs " + str(percent_women_smoker) + 
     " percent, for women. \nThis difference might be able to explain the gender difference better than discrimination.")

 
```
